# How to calibrate your neural network classifer:
## Getting accurate probabilities from your neural network model

In this tutorial we'll use the CIFAR-10 image dataset, and a classifier trained on it, to explore what model confidence calibration is, how we can measure it, and what methods we can put in place to rectify poorly calibrated models.

```python
# import relevant modules
import numpy as np
from keras.models import load_model
from keras.datasets import cifar10
import keras.utils
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
from sklearn.metrics import accuracy_score, log_loss
%matplotlib inline
```

```python
# Load the CIFAR10 data.
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

#### CIFAR-10
CIFAR-10 is a dataset containing a collection of images falling into 10 classes: airplanes, cars, birds, cats, deer, dogs, frogs, horses, ships, and trucks. It contains 60 000 low resolution (32x32) images and is often used to train ML and computer vision models. The low resolution allows for quick testing of different models.

```python
plt.imshow(x_train[112,])
plt.show()
```

First we need to do some preprocessing of the data to improve the accuracy.

```python
# Normalize data.
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# If subtract pixel mean from each image
x_train_mean = np.mean(x_train, axis=0)
x_train -= x_train_mean
x_test -= x_train_mean

y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

#### ResNet
Next step would be training the model but as we have limited time and resources, we will be using pre-trained [Keras ResNet model](https://keras.io/examples/cifar10_resnet/).

```python
# model = Model(<some layers>)
# ...
# model.fit(x_train, y_train)
# ...
# model.evaluate(x_test,y_test)
#
```

```python
model = load_model('cifar10_resnet.h5')
y_pred = model.predict(x_test)
#scores = model.evaluate(x_test, y_test)
```

```python
acc_score = accuracy_score(np.argmax(y_test,1),np.argmax(y_pred,1))
loss_score = log_loss(y_test, y_pred)
```

```python
print('Test accuracy is {0:.2f}, test loss is {1:.2f}'.format(acc_score, loss_score))
```

## 2. How can we measure calibration?

Suppose we consider our model as a binary classifier for illustration: either an image is in class 0, or it isn't. We can simplify our labels and predictions as follows:

```python
# Labels:
y_test_0 = [1 if y_val==0 else 0 for y_val in np.argmax(y_test, axis=1)]

# Predictions:
y_pred_0 = y_pred[:,0]
```

### 2.1 Reliability diagrams
Can you use calibration curve from scikit learn to show how calibrated the model is on our data? How would you interpret the results?

```python
### YOUR CODE HERE

```

```python
# calculate the values of calibration curve for bin 0 vs all
prob_true_0, prob_pred_0 = calibration_curve(y_test_0, y_pred_0, n_bins=10)

# Plot the calibration curve for ResNet in comparison with what a perfectly calibrated model would look like
fig = plt.figure(figsize=(10, 10))
plt.plot([0, 1], [0, 1], color="#FE4A49", linestyle=":", label="Perfectly calibrated model")
plt.plot(prob_pred_0, prob_true_0, "s-", label="ResNet (class 0 vs all)", color="#162B37")

plt.ylabel("Fraction of positives", fontsize=16)
plt.xlabel("Mean predicted value", fontsize=16)

plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True, color="#B2C7D9")

plt.tight_layout()
```

### 2.2 Expected calibration error
Given the explanation of ECE, can you calculate the error for our dataset?

```python
### YOUR CODE HERE 
```

```python
n_bins_0 = len(prob_true_0)
n_0 = len(y_pred_0)
pred_hist_0 = np.histogram(a=y_pred_0, range=(0, 1), bins=n_bins_0)[0] 

ece_0 = 0
for m in np.arange(n_bins_0):
    ece_0 = ece_0 + (pred_hist_0[m] / n_0) * np.abs(prob_true_0[m] - prob_pred_0[m])

ece_0
```

### 2.3 Maximum calibration error
Given the explanation of MCE, can you calculate it for our model?

```python
### YOUR CODE HERE 

```

```python
mce_0 = 0
for m in np.arange(n_bins_0):
    mce_0 = max(mce_0, np.abs(prob_true_0[m] - prob_pred_0[m]))

mce_0
```

### 2.4 Root mean square calibration error
Given the explanation, can you calculate RMSCE for our data?

```python
### YOUR CODE HERE 
```

```python
np.sqrt(np.sum(np.square((prob_true_0 - prob_pred_0)*100) * pred_hist_0) / np.sum(pred_hist_0))
```

### 2.5 Multiclass problems
Extending the definition of these metrics, we can use them for multiclass classifiers, too. Can you show the reliability diagrams and calculate the calibration errors for the 10 classes?

```python
### YOUR CODE HERE
```

```python
fig = plt.figure(figsize=(10, 10))

y_test_dict = {}
y_pred_dict = {}
prob_true = {}
prob_pred = {}
for i in range(10):
    y_test_dict[i] = [1 if y_val==i else 0 for y_val in np.argmax(y_test, axis=1)]

    # Predictions:
    y_pred_dict[i] = y_pred[:, i]

    # calculate the values of calibration curve for each bin
    prob_true[i], prob_pred[i] = calibration_curve(y_test_dict[i], y_pred_dict[i], n_bins=10)

    # Plot the calibration curve for each bin in comparison with what a perfectly calibrated model would look like
    plt.plot(prob_pred[i], prob_true[i], "s-", label="ResNet (class {0} vs all)".format(i))

plt.plot([0, 1], [0, 1], color="#162B37", linestyle=":", label="Perfectly calibrated model")
plt.ylabel("Fraction of positives", fontsize=16)
plt.xlabel("Mean predicted value", fontsize=16)

plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True, color="#B2C7D9")

plt.tight_layout()
```

```python
ece = []
mce = []
rmsce = []
for i in range(10):
    n_bins = len(prob_true[i])
    n = len(y_pred_dict[i])
    pred_hist = np.histogram(a=y_pred_dict[i], range=(0, 1), bins=n_bins)[0] 
    
    ece.append(0)
    mce.append(0)
    for m in np.arange(n_bins):
        ece[i] = ece[i] + (pred_hist[m] / n) * np.abs(prob_true[i][m] - prob_pred[i][m])
        mce[i] = max(mce[i], np.abs(prob_true[i][m] - prob_pred[i][m]))
        
    rmsce.append(np.sqrt(np.sum(np.square((prob_true[i] - prob_pred[i])*100) * pred_hist) / np.sum(pred_hist)))

    print("ECE for class {0} is {1}".format(i, ece[i]))
    print("MCE for class {0} is {1}".format(i, mce[i]))
    print("RMSCE for class {0} is {1}\n".format(i, rmsce[i]))
    

print("Overall ECE of ResNet is {0}".format(sum([0.1*val for val in ece])))
print("Overall MCE of ResNet is {0}".format(max(mce)))
print("Overall RMSCE of ResNet is {0}".format(sum([0.1*val for val in rmsce])))
```
